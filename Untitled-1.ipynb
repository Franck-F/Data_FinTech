```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Remplace par ta clé API Alpha Vantage
API_KEY = "M169Z5DSH2DNIL94"

# Liste des actifs à récupérer
assets = {
    "Bitcoin": "BTCUSD",
    "S&P 500": "SPX",
    "Or": "XAUUSD"
}

def fetch_data(symbol):
    """ Récupère les données de l'API Alpha Vantage """
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={API_KEY}"
    response = requests.get(url)
    data = response.json()

    if "Time Series (Daily)" not in data:
        print(f"Erreur de récupération pour {symbol}")
        return None

    # Transformer les données JSON en DataFrame
    df = pd.DataFrame.from_dict(data["Time Series (Daily)"], orient="index")
    df = df.rename(columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",
        "4. close": "Close",
        "5. volume": "Volume"
    }).astype(float)
    df.index = pd.to_datetime(df.index)  # Convertir en datetime
    df.sort_index(inplace=True)
    return df

# Récupération des données pour chaque actif
data_dict = {}
for asset_name, symbol in assets.items():
    print(f"Récupération des données pour {asset_name} ({symbol})...")
    data = fetch_data(symbol)
    if data is not None:
        data_dict[asset_name] = data

# Création du graphique avec un deuxième axe Y
fig, ax1 = plt.subplots(figsize=(10,5))

ax2 = ax1.twinx()  # Deuxième axe Y

for asset_name, df in data_dict.items():
    if asset_name == "Bitcoin":
        ax1.plot(df.index, df["Close"], label=asset_name, color="blue")
    elif asset_name == "Or":
        ax2.plot(df.index, df["Close"], label=asset_name, color="orange")

# Légendes et labels
ax1.set_xlabel("Date")
ax1.set_ylabel("Prix Bitcoin (USD)", color="blue")
ax2.set_ylabel("Prix Or (USD)", color="orange")

ax1.set_title("Comparaison des prix de Bitcoin et Or avec double échelle")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
ax1.grid()

plt.show()
```

```python
def compute_macd(df):
    """ Calcule l'indicateur MACD et sa ligne de signal """
    df["EMA_12"] = df["Close"].ewm(span=12, adjust=False).mean()
    df["EMA_26"] = df["Close"].ewm(span=26, adjust=False).mean()
    df["MACD"] = df["EMA_12"] - df["EMA_26"]
    df["Signal"] = df["MACD"].ewm(span=9, adjust=False).mean()
    return df
```

```python
data_dict["Bitcoin"] = compute_macd(data_dict["Bitcoin"])
```

```python
plt.figure(figsize=(10,5))
plt.plot(data_dict["Bitcoin"].index, data_dict["Bitcoin"]["MACD"], label="MACD", color="blue")
plt.plot(data_dict["Bitcoin"].index, data_dict["Bitcoin"]["Signal"], label="Signal", color="red")
plt.axhline(0, color='black', linestyle='--', linewidth=0.5)  # Ligne zéro pour repère
plt.title("MACD de Bitcoin")
plt.xlabel("Date")
plt.ylabel("Valeur MACD")
plt.legend()
plt.grid()
plt.show()
```

```python
def compute_bollinger(df):
    """ Calcule les bandes de Bollinger """
    df["SMA_20"] = df["Close"].rolling(window=20).mean()
    df["Bollinger_High"] = df["SMA_20"] + (2 * df["Close"].rolling(window=20).std())
    df["Bollinger_Low"] = df["SMA_20"] - (2 * df["Close"].rolling(window=20).std())
    return df
```

```python
data_dict["Bitcoin"] = compute_bollinger(data_dict["Bitcoin"])
```

```python
plt.figure(figsize=(10,5))
plt.plot(data_dict["Bitcoin"].index, data_dict["Bitcoin"]["Close"], label="Prix de clôture", color="blue")
plt.plot(data_dict["Bitcoin"].index, data_dict["Bitcoin"]["Bollinger_High"], label="Bande Supérieure", color="red", linestyle="--")
plt.plot(data_dict["Bitcoin"].index, data_dict["Bitcoin"]["Bollinger_Low"], label="Bande Inférieure", color="green", linestyle="--")
plt.fill_between(data_dict["Bitcoin"].index, data_dict["Bitcoin"]["Bollinger_Low"], data_dict["Bitcoin"]["Bollinger_High"], color='gray', alpha=0.1)
plt.title("Bandes de Bollinger de Bitcoin")
plt.xlabel("Date")
plt.ylabel("Prix")
plt.legend()
plt.grid()
plt.show()
```

Calculer la matrice de corrélation entre les actifs

```python
import seaborn as sns

# Vérifier que toutes les données sont disponibles avant d'appliquer la corrélation
assets_to_include = ["Bitcoin", "Or", "S&P 500"]
correlation_data = pd.DataFrame()

for asset in assets_to_include:
    if asset in data_dict:
        correlation_data[asset] = data_dict[asset]["Close"]

# Calculer la matrice de corrélation
correlation_matrix = correlation_data.corr()

# Affichage avec Seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de Corrélation entre Bitcoin, Or et S&P 500")
plt.show()
```
